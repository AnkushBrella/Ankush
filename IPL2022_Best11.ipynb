#### Sorting players according to their stats

```python
#Sorting the best batsmans according to the following stats : Runs, Strike_rate, Average, Matches_played.

Bat_run=run_data.sort_values(by = "Runs", ascending = False)
top_bat_run = Bat_run.head(20)

Bat_Mat=run_data.sort_values(by = "Mat", ascending = False)
Bat_Inns=Bat_Mat[Bat_Mat["Inns"]>=10]

Bat_avg=Bat_Inns.sort_values(by = "Avg", ascending = False)
top_bat_avg = Bat_avg.head(20)

Bat_sr=Bat_Inns.sort_values(by = "SR", ascending = False)
top_bat_sr = Bat_sr.head(20)
```

```python
#Sorting the best bowlers according to the following stats : Wickets, Economy, Average, Matches_played.

Ball_Wkts=wicket_data.sort_values(by = "Wkts", ascending = False)
top_ball_wkts = Ball_Wkts.head(20)

Bowl_Mat=wicket_data.sort_values(by = "Mat", ascending = False)
Bowl_Inns=Bowl_Mat[Bowl_Mat["Inns"]>=10]

Bowl_Avg=Bowl_Inns.sort_values(by = "Avg")
top_bowl_avg = Bowl_Avg.head(20)

Bowl_Econ =Bowl_Inns.sort_values(by = "Econ")
top_bowl_econ = Bowl_Econ.head(20)
```

```python
#Sorting the best Wicket_keeper according to the following stats : Runs, Strike_rate, Average, Dismissals.

WK_run=keeper_data.sort_values(by = "Runs", ascending = False)
top_wk_run = WK_run.head()

WK_Dismissals =keeper_data.sort_values(by = "Dismissals", ascending = False)
top_wk_dismissals = WK_Dismissals.head()

WK_Avg =keeper_data.sort_values(by = "Avg", ascending = False)
top_wk_avg = WK_Avg.head()

WK_SR =keeper_data.sort_values(by = "SR", ascending = False)
top_wk_sr = WK_SR.head()
```

```python
#Sorting the best batsmen according to the following stats : Runs, Strike_rate, Average, Wickets, Economy, Matches_played.

allrounder_run= allrounder_data.sort_values(by="Runs", ascending = False)
top_alrd_run = allrounder_run.head(10)

allrounder_wkt= allrounder_data.sort_values(by="Wkts", ascending = False)
top_alrd_wkt = allrounder_wkt.head(10)

allrounder_Mat = allrounder_data.sort_values(by = "Mat", ascending = False)
allrounder_inns = allrounder_Mat[allrounder_Mat["Inns"]>=10]

allrounder_SR =allrounder_inns.sort_values(by = "SR", ascending = False)
top_alrd_sr = allrounder_SR.head(10)

allrounder_Econ =allrounder_inns.sort_values(by = "Econ",)
top_alrd_econ = allrounder_Econ.head(10)

allrounder_avg =allrounder_inns.sort_values(by = "Avg", ascending = False)
top_alrd_avg = allrounder_avg.head(10)
```

# Final best 11 of the campaign:

```python
#assigning positions to the players

batter1 = run_data.loc[(run_data["Player"] == 'Jos Buttler')]
batter2 = run_data.loc[(run_data["Player"] == 'K L Rahul')]
batter3 = run_data.loc[(run_data["Player"] == 'Shubman Gill')]
batter4 = run_data.loc[(run_data["Player"] == 'David Miller')]
keeper1 = run_data.loc[(run_data["Player"] == 'Dinesh Karthik')]

allrounder1 = allrounder_data.loc[(allrounder_data["Player"] == 'Hardik Pandya')]
allrounder2 = allrounder_data.loc[(allrounder_data["Player"] == 'Andre Russell')]
allrounder3 = allrounder_data.loc[(allrounder_data["Player"] == 'Axar Patel')]

bowler1 = wicket_data.loc[(wicket_data["Player"] == 'Kagiso Rabada')]
bowler2 = wicket_data.loc[(wicket_data["Player"] == 'Umran Malik')]
bowler3 = wicket_data.loc[(wicket_data["Player"] == 'Yuzvendra Chahal')]
```

```python
#droping the labels from the list

final_bat = [batter1, batter2, batter3, batter4, keeper1]
final_team_bat = pd.concat(final_bat)
final_team_bat = final_team_bat.drop(labels=['Mat','Inns','NO','Runs','HS','Avg','SR','100','50','4s','6s'],axis=1)

final_all = [allrounder1, allrounder2, allrounder3]
final_team_all = pd.concat(final_all)
final_team_all = final_team_all.drop(labels=['Mat','Inns','NO','Runs','HS','Avg','SR','100','50','4s','6s',
                                             'Ov','Wkts','Ball_Avg','Econ','SR.1','4w','5w'],axis=1)

final_ball = [bowler1, bowler2, bowler3]
final_team_ball = pd.concat(final_ball)
final_team_ball = final_team_ball.drop(labels=['Mat','Inns','Ov','Runs','Wkts','Avg','Econ','SR','4w','5w'],axis=1)
```

```python
final_team = pd.concat([final_team_bat,final_team_all,final_team_ball], ignore_index =True)
```
